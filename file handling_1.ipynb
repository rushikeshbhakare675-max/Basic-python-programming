```python
# File handling in Python is the process of creating, reading, writing, updating, 
# and deleting files using functions like open() 
# and file modes.
```

```python
# create a simple file
```

```python
with open("sample.txt","w") as f:
     f.write("name: rushi\n")
     f.write("Role: Data scientist\n")
     f.write("location: India\n")
```

```python
# reading the file
```

```python
with open("sample.txt", "r") as f:
    data = f.read()
    print(data)
```

```python
# riding line by line
```

```python
with open("sample.txt", "r") as f:
    for line in f:
        print(line.strip())
```
