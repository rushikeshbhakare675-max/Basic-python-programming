```python
#7.	Write a Python program to generate the Fibonacci series, calculate its sum, and display the largest term generated.
```

```python
n = int(input("Enter the number of terms: "))
a, b = 0, 1
fib_series = []
sum_series = 0
for i in range(n):
    fib_series.append(a)
    sum_series += a
    a, b = b, a + b
print("Fibonacci Series:", fib_series)
print("Sum of the series:", sum_series)

if fib_series:
    print("Largest term generated:", max(fib_series))
else:
    print("No terms generated.")
```

```python
#11.	Write a Python program to reverse a string, count the vowels, and count the consonants.
```

```python
text = input("Enter a string: ")


reversed_text = text[::-1]

vowels = 0
consonants = 0
for char in text.lower():
    if char.isalpha():
        if char in "aeiou":
            vowels += 1
        else:
            consonants += 1

print("Reversed String:", reversed_text)
print("Number of Vowels:", vowels)
print("Number of Consonants:", consonants)
```

```python
#12.	Write a Python program to check whether a string is a palindrome, count the number of words, and convert it to uppercase.
```

```python
text = input("Enter a string: ")


if text.lower() == text[::-1].lower():
    palindrome = "Palindrome"
else:
    palindrome = "Not a Palindrome"
word_count = len(text.split())


uppercase_text = text.upper()


print("Palindrome Check:", palindrome)
print("Number of Words:", word_count)
print("Uppercase String:", uppercase_text)
```

```python
#16.	Write a Python program to create a nested list, access its elements, and calculate the sum of all numeric values.
```

```python
nested_list = [
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
]

print("First row:", nested_list[0])
print("Element at row 2, column 3:", nested_list[1][2])

total_sum = 0

for row in nested_list:
    for value in row:
        total_sum += value

print("Nested List:", nested_list)
print("Sum of all numeric values:", total_sum)
```

```python
import math

num = int(input("Enter a number: "))
base = int(input("Enter base value: "))
exponent = int(input("Enter exponent value: "))
sqrt_result = math.sqrt(num)

power_result = math.pow(base, exponent)


factorial_result = math.factorial(num)
print("Square root of", num, "=", sqrt_result)
print(base, "raised to the power", exponent, "=", power_result)
print("Factorial of", num, "=", factorial_result)
```

```python
#24.	Write a Python program to create a matrix using nested lists, display the matrix, and calculate row-wise sums.
```

```python
matrix = [
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
]


print("Matrix:")
for row in matrix:
    print(row)


print("\nRow-wise Sums:")
for i in range(len(matrix)):
    row_sum = sum(matrix[i])
    print(f"Sum of row {i+1}:", row_sum)
```

```python
#25.	Write a Python program to accept N numbers from the user, store them in a list, and display the maximum, minimum, and average values.

```

```python
n = int(input("Enter how many numbers you want to enter: "))

numbers = []
for i in range(n):
    num = float(input(f"Enter number {i+1}: "))
    numbers.append(num)


maximum = max(numbers)
minimum = min(numbers)
average = sum(numbers) / n


print("\nList:", numbers)
print("Maximum value:", maximum)
print("Minimum value:", minimum)
print("Average value:", average)
```
